7) How memory is managed in Python?
-->Python handles memory management automatically through a private heap and specialized mechanisms, so developers don't have to manually allocate or free space.
-Private Heap: All Python objects and data structures are stored here.
-Reference Counting: Every object tracks how many references point to it. When this count drops to zero, the object is immediately deallocated.
-Garbage Collection: Python uses a generational garbage collector to find and clean up "reference cycles" that reference counting alone can't catch.

```python
#29) Write a Python function to get the largest number, smallest num and sum of all from a list.
def list_stats(lst):
    largest = max(lst)
    smallest = min(lst)
    total = sum(lst)
    
    return largest, smallest, total

numbers = [10, 5, 20, 8]

l, s, t = list_stats(numbers)

print("Largest:", l)
print("Smallest:", s)
print("Sum:", t)
```

```python
#61)Write a Python function to calculate the factorial of a number (anonnegative integer)
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
#63)Write a Python function to check whether a number is perfect or not. 
def is_perfect(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total == n

print(is_perfect(7))
```
